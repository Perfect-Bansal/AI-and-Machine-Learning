```python
#Python program that converts between different units of measurement.
def convert_units(conversion_type, value, direction):
    """
    Converts units based on the type and direction selected.

    Parameters:
    conversion_type (str): Type of conversion ('length', 'weight', 'volume')
    value (float): The numeric value to convert
    direction (str): Specific conversion direction

    Returns:
    float: Converted value
    """

    if conversion_type == "length":
        if direction == "m_to_ft":
            return value * 3.28084
        elif direction == "ft_to_m":
            return value / 3.28084

    elif conversion_type == "weight":
        if direction == "kg_to_lbs":
            return value * 2.20462
        elif direction == "lbs_to_kg":
            return value / 2.20462

    elif conversion_type == "volume":
        if direction == "l_to_gal":
            return value * 0.264172
        elif direction == "gal_to_l":
            return value / 0.264172

    else:
        return None


def main():
    try:
        print("Select conversion type: length / weight / volume")
        conversion_type = input("Enter type: ").lower()

        value = float(input("Enter value to convert: "))

        print("Enter conversion direction:")
        print("Length: m_to_ft / ft_to_m")
        print("Weight: kg_to_lbs / lbs_to_kg")
        print("Volume: l_to_gal / gal_to_l")

        direction = input("Enter direction: ").lower()

        result = convert_units(conversion_type, value, direction)

        if result is None:
            print("Unsupported conversion type or direction.")
        else:
            print("Converted value:", round(result, 4))

    except ValueError:
        print("Invalid input! Please enter a numeric value.")


if __name__ == "__main__":
    main()
```

```python
#Python program that performs various mathematical operations on a list of numbers.
def find_sum(numbers):
    """
    Returns the sum of a list of numbers.

    Parameters:
    numbers (list): List of numeric values

    Returns:
    float: Sum of numbers
    """
    return sum(numbers)


def find_average(numbers):
    """
    Returns the average of a list of numbers.

    Parameters:
    numbers (list): List of numeric values

    Returns:
    float: Average of numbers
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Returns the maximum value in a list.

    Parameters:
    numbers (list): List of numeric values

    Returns:
    float: Maximum value
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Returns the minimum value in a list.

    Parameters:
    numbers (list): List of numeric values

    Returns:
    float: Minimum value
    """
    return min(numbers)


def main():
    try:
        print("Choose operation: sum / average / max / min")
        operation = input("Enter operation: ").lower()

        user_input = input("Enter numbers separated by spaces: ")
        numbers = list(map(float, user_input.split()))

        if not numbers:
            raise ValueError("Empty list is not allowed.")

        if operation == "sum":
            print("Result:", find_sum(numbers))
        elif operation == "average":
            print("Result:", find_average(numbers))
        elif operation == "max":
            print("Result:", find_maximum(numbers))
        elif operation == "min":
            print("Result:", find_minimum(numbers))
        else:
            print("Invalid operation selected.")

    except ValueError:
        print("Invalid input! Please enter valid numbers.")


if __name__ == "__main__":
    main()
```

```python
# 4.2 Exercise on List Manipulation
# 1. Extract Every Other Element
def extract_every_other(lst):
    """Returns every other element starting from first."""
    return lst[::2]


# 2. Get Sublist (inclusive)
def get_sublist(lst, start, end):
    """Returns sublist from start to end (inclusive)."""
    return lst[start:end+1]


# 3. Reverse List
def reverse_list(lst):
    """Returns reversed list using slicing."""
    return lst[::-1]


# 4. Remove First and Last
def remove_first_last(lst):
    """Removes first and last elements."""
    return lst[1:-1]


# 5. Get First n Elements
def get_first_n(lst, n):
    """Returns first n elements."""
    return lst[:n]


# 6. Get Last n Elements
def get_last_n(lst, n):
    """Returns last n elements."""
    return lst[-n:]


# 7. Reverse Skip
def reverse_skip(lst):
    """Returns every second element starting from second-last, backwards."""
    return lst[-2::-2]
```

```python
#  4.3 Exercise on Nested List
# 1. Flatten Nested List (Simple)
def flatten(lst):
    """Flattens one level nested list."""
    result = []
    for item in lst:
        for element in item:
            result.append(element)
    return result
# 2. Access Nested Element
def access_nested_element(lst, indices):
    """
    Returns element from nested list using list of indices.
    """
    element = lst
    for index in indices:
        element = element[index]
    return element
# 3. Sum Nested (Any Depth)
def sum_nested(lst):
    """Returns sum of all elements in nested list."""
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total
# 4. Remove Element from Nested List
def remove_element(lst, elem):
    """Removes all occurrences of elem from nested list."""
    new_list = []
    for item in lst:
        if isinstance(item, list):
            new_list.append(remove_element(item, elem))
        elif item != elem:
            new_list.append(item)
    return new_list
# 5. Find Maximum in Nested List
def find_max(lst):
    """Finds maximum element in nested list."""
    max_value = float('-inf')
    for item in lst:
        if isinstance(item, list):
            max_value = max(max_value, find_max(item))
        else:
            max_value = max(max_value, item)
    return max_value
# 6. Count Occurrences in Nested List
def count_occurrences(lst, elem):
    """Counts occurrences of elem in nested list."""
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count
# 7. Deep Flatten (Any Depth)
def deep_flatten(lst):
    """Flattens deeply nested list."""
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result
# 8. Average of Nested List
def average_nested(lst):
    """Returns average of nested list elements."""
    flat_list = deep_flatten(lst)
    return sum(flat_list) / len(flat_list)
```
